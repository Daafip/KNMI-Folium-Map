# Scrape KNMI values

```python
knmi_daily = "https://www.knmi.nl/nederland-nu/klimatologie/daggegevens"
```

Get data:

```python
page = urlopen(knmi_daily)
html_bytes = page.read()
html = html_bytes.decode("utf-8")
```

select only the wanted table:

```python
table = html[html.find('<table  class="table">'):html.find('</table>')]
table_data = [[cell.text.strip() for cell in row("td")]
                         for row in BeautifulSoup(table)("tr")]
headers = [[cell.text.strip() for cell in row("th")]
                      for row in BeautifulSoup(table)("tr")][0]
```

remove the unwanted rows

```python
df_stations_daily = pd.DataFrame(data=table_data, columns=headers)
df_stations_daily.replace({"":pd.NA},inplace=True)
df_stations_daily.dropna(inplace=True)
```

# [KNMI daily data](https://www.knmi.nl/nederland-nu/klimatologie/daggegevens)

```python
date_cols = ["Vanaf", "Tot en met"]
# df_stations_daily = pd.read_excel("knmi_daily_station_data.xlsx", parse_dates=date_cols).dropna()
# fix dates to be readable
for col in date_cols:
    df_stations_daily[col] = pd.to_datetime(df_stations_daily[col])
    df_stations_daily[col] = df_stations_daily[col].apply(lambda x:f'{str(x)[:10]}')

df_stations_daily["Nummer"] = df_stations_daily["Nummer"].astype(int)
df_stations_daily["URL"] = df_stations_daily["Nummer"].apply(lambda x: f"<a href=https://cdn.knmi.nl/knmi/map" + \
                            f"/page/klimatologie/gegevens/daggegevens/etmgeg_{x}.zip>Download KNMI daily data</a>")

# add nederland to adress
df_stations_daily["Station"].replace({"Texelhors": "Den Hoorn, Texel"},inplace=True)
df_stations_daily["Station"] = df_stations_daily["Station"].apply(lambda x: f"{x}, Nederland")
df_stations_daily.head()
```

make into geodata frame & geocode, this takes a sec to locate the stations

```python
gdf_knmi_daily = gpd.GeoDataFrame(data=df_stations_daily)
```

```python
gdf_adresses_daily = gpd.tools.geocode(gdf_knmi_daily.Station)
```

geocoding creates a new df, add this into the old

```python
df_stations_daily = gdf_knmi_daily.copy()
for col in gdf_adresses_daily.columns:
    df_stations_daily[col] = gdf_adresses_daily[col]
```

Panda's opperations make it into a df, so remake it into a df, then export for the map & add

```python
gdf_stations_daily = gpd.GeoDataFrame(data=df_stations_daily, geometry=df_stations_daily.geometry)
```

```python
gdf_stations_daily = gdf_stations_daily.to_crs(epsg=4326)
```

```python
icon = folium.features.Icon(color="blue",icon="none")

gjson = folium.features.GeoJson(
                                gdf_stations_daily,
                                marker = folium.features.Marker(icon=icon)
                                ).add_to(nl_map)

# create
popup = folium.features.GeoJsonPopup(
    fields=['Nummer', 'Station', 'Vanaf', 'Tot en met','URL'],
    aliases=['Station nummer', "Plek naam", "Dagelijke data vanaf", "Dagelijkse data tot (laaste update kaart)", "URL"]    

).add_to(gjson)
```

# Hourly

### [KNMI hourly data](https://www.knmi.nl/nederland-nu/klimatologie/uurgegevens)

```python
knmi_hourly_url = "https://www.knmi.nl/nederland-nu/klimatologie/uurgegevens"
```

```python
page = urlopen(knmi_hourly_url)
html_bytes = page.read()
html = html_bytes.decode("utf-8")
```

```python
def find_all(a_str, sub):
    """returns all instances in a string, thanks to:
    https://stackoverflow.com/questions/4664850/how-to-find-all-occurrences-of-a-substring"""
    start = 0
    while True:
        start = a_str.find(sub, start)
        if start == -1: return
        yield start
        start += len(sub) # use start += 1 to find overlapping matches
```

Hourly data is less 'organised' for scraping, so we do it ourselves

```python
start = html.find('<table  class="table">')
end = html.find('</table>')
lst_data = []
for start, end in zip(list(find_all(html, '<table  class="table">')), list(find_all(html, '</table>'))):
    table = html[start:end]
    table_data = [[cell.text for cell in row("td")]
                             for row in BeautifulSoup(table)("tr")]
    headers = [[cell.text for cell in row("th")]
                          for row in BeautifulSoup(table)("tr")][0]
    headers = [header.strip() for header in headers[0].split('-')]
    if len(headers) > 2:
        headers = [headers[0], headers[1]+'-'+headers[2]]
    dates_availible = [date.strip().replace(" ","") for date in table_data[1][0].split('|')[1:] if len(date.strip()) > 1]
    base = "<a href =https://cdn.knmi.nl/knmi/map/page/klimatologie/gegevens/uurgegevens/uurgeg"
    station = headers[0].strip()
    link_downloads = [f'{base}_{station}_{date}.zip>{date}</a>' for date in dates_availible]
    total_link = ""
    for i, link in enumerate(link_downloads):
        total_link += link
        if i != len(link_downloads):
            total_link += "\n"
    # only take actual stations, not "all stations"
    if len(headers) > 1:
        lst_data.append(headers + [total_link])
```

```python
df_stations_hourly = pd.DataFrame(data=lst_data, columns=["Nummer","Station","Links"])

df_stations_hourly["Station"].replace({"Texelhors": "Den Hoorn, Texel"},inplace=True)
df_stations_hourly["Station"] = df_stations_hourly["Station"].apply(lambda x: f"{x}, Nederland")
```

```python
gdf_hourly = gpd.GeoDataFrame(data=df_stations_hourly)
```

```python
gdf_adresses_hourly = gpd.tools.geocode(gdf_hourly.Station)
```

```python
df_stations_hourly = gdf_hourly.copy()
for col in gdf_adresses_hourly.columns:
    df_stations_hourly[col] = gdf_adresses_hourly[col]
```

```python
gdf_stations_hourly = gpd.GeoDataFrame(data=df_stations_hourly, geometry=df_stations_hourly.geometry)
```

```python
gdf_stations_hourly = gdf_stations_hourly.to_crs(epsg=4326)
```

```python
# Create map
control_bool = True
nl_map_hourly = folium.Map(
                    location        = [NL_outline.iloc[0].geometry.centroid.y, NL_outline.iloc[0].geometry.centroid.x],
                    zoom_start      = 8,
                    zoom_control    = control_bool,
                    scrollWheelZoom = control_bool,
                    dragging        = control_bool
                    )

NL_outline_4326 = NL_outline.to_crs(epsg=4326)
# add outline
folium.features.GeoJson(NL_outline_4326).add_to(nl_map_hourly)

# change icon, but couldnt find a meaning full one so none https://getbootstrap.com/docs/3.3/components/
icon = folium.features.Icon(color="red",icon="none")

gjson = folium.features.GeoJson(
                                gdf_stations_hourly,
                                marker = folium.features.Marker(icon=icon)
                                ).add_to(nl_map_hourly)

# create
popup = folium.features.GeoJsonPopup(
    fields=['Nummer', 'Station', 'Links'],
    aliases=['Station nummer', "Plek naam", "Download Hourly data"]    
).add_to(gjson)
```
